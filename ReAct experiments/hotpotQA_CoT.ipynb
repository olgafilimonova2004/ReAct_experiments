# Use

```python
def ask_LLM(message :str) -> str:
  input_ids = tokenizer.apply_chat_template([
    {'role': 'user', 'content': message}
  ], add_generation_prompt=True, tokenize=True, return_tensors='pt')
  input_ids = input_ids.to(model.device)

  output = model.generate(
      input_ids,
      generation_config=generation_config
  )[:, input_ids.shape[-1]:].detach().cpu()

  output = tokenizer.batch_decode(output, skip_special_tokens=True)[0]
  return output
```

```python
def CoT (idx, question):
  example="""User: Here is an example of how to answer the question: Which is located in India, University of the Philippines or The Northcap University?
  Write the short final answer in the last line (not the sentence, just the answer).
  Assistant: Let’s think step by step:
  1. Identify the countries associated with each university:
    - 'University of the Philippines' has 'Philippines' in its name, suggesting it's located in the Philippines.
    - 'The Northcap University' does not include a country name, but a quick search shows it's based in Gurugram, India.
  2. Compare with the target country (India):
    - The Northcap University is in India.
    - The University of the Philippines is not.
  3. Final Answer: The Northcap University
  The Northcap University
          """
  prompt=example+f'\nQuestion: {question}'
  pred=ask_LLM(prompt)
  print(f"\n{idx} Question: {question}\nAssistant's answer: {pred}")
  return pred
```

```python
import json
import os
import numpy as np
import re
import string
from collections import Counter

def normalize_answer(s):
  def remove_articles(text):
    return re.sub(r"\b(a|an|the)\b", " ", text)

  def white_space_fix(text):
      return " ".join(text.split())

  def remove_punc(text):
      exclude = set(string.punctuation)
      return "".join(ch for ch in text if ch not in exclude)

  def lower(text):
      return text.lower()

  return white_space_fix(remove_articles(remove_punc(lower(s))))

def f1_score(prediction, ground_truth):
  normalized_prediction = normalize_answer(prediction)
  normalized_ground_truth = normalize_answer(ground_truth)

  ZERO_METRIC = (0, 0, 0)

  if normalized_prediction in ['yes', 'no', 'noanswer'] and normalized_prediction != normalized_ground_truth:
    return ZERO_METRIC
  if normalized_ground_truth in ['yes', 'no', 'noanswer'] and normalized_prediction != normalized_ground_truth:
    return ZERO_METRIC

  prediction_tokens = normalized_prediction.split()
  ground_truth_tokens = normalized_ground_truth.split()
  common = Counter(prediction_tokens) & Counter(ground_truth_tokens)
  num_same = sum(common.values())
  if num_same == 0:
    return ZERO_METRIC
  precision = 1.0 * num_same / len(prediction_tokens)
  recall = 1.0 * num_same / len(ground_truth_tokens)
  f1 = (2 * precision * recall) / (precision + recall)
  return f1, precision, recall

def get_metrics(pred, answer):
    if answer is not None:
        pred = normalize_answer(pred) #(self.data[self.data_idx][1])
        gt = normalize_answer(answer)
        em = (pred == gt)
        f1 = f1_score(pred, gt)[0]
        return {'answer': pred, 'gt_answer': gt, 'reward': em, 'em': em, 'f1': f1}
    return {'answer': pred, 'gt_answer': gt, 'reward': 0, 'em': 0, 'f1': 0}
```
