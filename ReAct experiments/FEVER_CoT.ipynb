# Use

```python
def ask_LLM(message :str) -> str:
  input_ids = tokenizer.apply_chat_template([
    {'role': 'user', 'content': message}
  ], add_generation_prompt=True, tokenize=True, return_tensors='pt')
  input_ids = input_ids.to(model.device)

  output = model.generate(
      input_ids,
      generation_config=generation_config
  )[:, input_ids.shape[-1]:].detach().cpu()

  output = tokenizer.batch_decode(output, skip_special_tokens=True)[0]
  return output
```

```python
def CoT (idx, question, stop='\n'):
  shots="Determine if there is Observation that SUPPORTS or REFUTES a Claim, or if there is NOT ENOUGH INFORMATION. Write the final answer in format finish[] as shown below.\nClaim: Nikolaj Coster-Waldau worked with the Fox Broadcasting Company.\nThought: Nikolaj William Coster-Waldau appeared in the 2009 Fox television film Virtuality, so he has worked with the Fox Broadcasting Company.\nAnswer: finish[SUPPORTS]\n\nClaim: Stranger Things is set in Bloomington, Indiana.\nThought: Stranger Things is in the fictional town of Hawkins, Indiana, not in Bloomington, Indiana.\nAnswer: finish[REFUTES]\n\nClaim: Beautiful reached number two on the Billboard Hot 100 in 2003.?\nThought: The song peaked at number two on the Billboard Hot 100 in the United States, but not sure if it was in 2003.\nAnswer: finish[NOT ENOUGH INFO]\n"
  prompt=shots+question
  pred=ask_LLM(prompt)
  print(f"\n{idx} {question}\nAssistant's answer: {pred}")
  return pred
```
